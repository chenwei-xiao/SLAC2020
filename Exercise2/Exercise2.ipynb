# Part 1: Regularisation of Linear Models

In the lecture, you have familiarised yourself with Linear Models. A Linear Regression learns functions of the form

> $\pmb{Y} = f(\pmb{X}) + \epsilon$ 

where $x_{1:n}$ are usually the observations, $y$ is the variable we want to predict, and $\beta_{1:n}$ are the linear coefficients. In it's simplest, two-dimensional form, you might remember your high school algebra:

> $Y = \beta_0 + \beta_1 x + \epsilon$ with $\beta_0$ as *intercept* and $\beta_1$ as *slope*


Finding the best values for these linear coefficients or *weights* is what happens when we fit a Linear Model to data. 

In the last tutorial, you also learned about the Bias-Variance tradeoff and the danger of overfitting your model. You learned that careful selection of hyperparameters (like the $k$, in the knn model) is of great importance for proper model fitting. The Linear Model as described above, however, does not have any hyperparameters. However, as you have learned in the lecture, it is still possible to overfit with the Linear Model. To *regularize* a linear model, one can impose a penalty on the weights for getting to large. 

The Ridge regression and the Lasso regression are two linear models that impose a penalty on weights growth in their formulation. To the ordinary least squares minimisation of model error in an unregularised linear model

> $\hat{\beta}^{OLS} = \text{argmin}_\beta \{\underbrace{\sum_{i=1}^N(y_i-\beta_0-\sum_{j=1}^p x_{ij}\beta_j)^2}_{\text{RSS}}\} = \text{argmin}_\beta\{\underbrace{(\pmb{y} - \pmb{X} \beta)^T (\pmb{y} - \pmb{X} \beta)}_{\text{RSS}}\}$, where $\pmb{X} \beta$ is the model prediction (including the intercept $\beta_0$) and $\pmb{y}$ is the true value vector.

we add a term favoring smaller weights
 
> Ridge: $\hat{\beta}^{Ridge} = \text{argmin}_\beta\{\text{RSS} + \underbrace{\lambda\sum_{j=1}^p \beta_j^2}_{\text{Ridge penalty term}}\} $ 

> Lasso: $\hat{\beta}^{Lasso} = \text{argmin}_\beta\{\text{RSS} + \underbrace{\lambda\sum_{j=1}^p | \beta_j |}_{\text{Lasso penalty term}}\} $ 

## Loading the data

Consider a ’classical’ real-world statistical learning problem: Given a set of winter monthly sea level pressure (SLP) anomalies `X`, we want to predict associated winter monthly anomalies in precipitation (TOTPREC), which we call `Y`.

We would like to derive a regularized model for winter precipitation at a specific location (here: Zurich, Switzerland, 47.4$^\circ$N,8.5$^\circ$E) using the entire European sea level pressure anomaly field as a predictor. Hence, we use a dataset that has the monthly winter precipitation over Zurich, Switzerland as the local `Y` time series, and a $p$ x $n$ matrix `X` as predictors of the European sea level pressure anomaly field (with $p = 357$ as number of predictors$^1$ and $n = 201$ (of in total $3000$) as number of observations).

($^1$ The Central European area selected consists of a grid of $17 * 21 = 357$ grid points at the model solution of 2.5$^\circ$ degree)

We open the NetCDF file, extract the two variables in lists and close the NetCDF file again:

```r
## Define path to NetCDF file and open the file:
nc_path <- "data/cesm122_1000y_piC_DJF_anom.nc"
#nc_path <- "/teaching/2020_HS/SLAC/Data/Data_Exercises/cesm122_1000y_piC.nc"
nc_file <- nc_open(nc_path,write=F)

## Define which variables should be read from NetCDF file:
var1_read <- "PSL"
var2_read <- "TOTPREC"

## Create list object with variable data (dat), attributes (att),
## and the respective coordiantes:
PSL     <- list(dat=(ncvar_get(nc_file, var1_read)),
                att=(ncatt_get(nc_file, var1_read)),
                coord_lat=nc_file$dim$lat,
                coord_lon=nc_file$dim$lon,
                coord_time=nc_file$dim$time)

TOTPREC <- list(dat=(ncvar_get(nc_file, var2_read)),
                att=(ncatt_get(nc_file, var2_read)),
                coord_lat=nc_file$dim$lat,
                coord_lon=nc_file$dim$lon,
                coord_time=nc_file$dim$time)

## Close the NetCDF file:
nc_close(nc_file)
```

From the NetCDF file, we create a 2D-dataframe that contains the precipitation over Zürich `Precip_Zurich` and the sea level pressure anomalies across Europe (`PSL_-10.00_31.26`, `PSL_-7.50_31.26`, ...):

```r
coord_grid <- expand.grid(lon=TOTPREC$coord_lon$vals,lat=TOTPREC$coord_lat$vals)

## Zurich coordinates:
ZH_lon <- 8.5417
ZH_lat <- 47.4

## Get Zurich grid point coordinates and indices:
ZH_lon_gp <- TOTPREC$coord_lon$vals[ZH_ind_lon <- which.min(abs(TOTPREC$coord_lon$vals-ZH_lon))]
ZH_lat_gp <- TOTPREC$coord_lat$vals[ZH_ind_lat <- which.min(abs(TOTPREC$coord_lat$vals-ZH_lat))]

## Create new 2D dataframe with dependent variable Y (precipitation in Zurich, "Precip_Zurich")
## and explanatory variables / features / predictors X_i (PSL accross Europe,
## "PSL_-10.00_31.26", "PSL_-7.50_31.26", ...):
reg_dat <- data.frame(Precip_Zurich=TOTPREC$dat[ZH_ind_lon, ZH_ind_lat,]*3600*24*1000,
                      apply(expand.grid(lon_ind=1:dim(PSL$dat)[1],
                                        lat_ind=1:dim(PSL$dat)[2]),
                            1,function(x) PSL$dat[x[1],x[2],]))
                            
## Add coordinate information to PSL predictors:
names(reg_dat)[2:ncol(reg_dat)] <- apply(coord_grid,1,function(x) paste("PSL",sprintf("%.2f",x[1]),
                                                                        sprintf("%.2f",x[2]),sep="_"))
str(reg_dat[,1:5])
                                         
## Put exact coordinates and data into a new list:
reg_data <- list(dat=reg_dat, PSL_coord=coord_grid)
```

## Fitting a Ridge/Lasso Regression

Now we move to the actual exercise. We want to explain `Y` (precipitation in Zürich) by `X` (PSL map of europe). To make the effect of regularisation visible, we only use 201 months for our model. 171 of them are used for training, and 30 are used for testing.

Define `X` and `Y` values for learning

```r
sel_months <- 800:1000    ## We only use 201 months 
length_test_dataset <- 30 ## 30 months of 201 months left for test dataset

X = as.matrix(reg_data$dat[,-1])[sel_months,] ## -1 index means without first column, i.e. not precipitation but only PSL
Y = as.matrix(reg_data$dat$Precip_Zurich[sel_months])
```

We define the indices for testing ...

```r
test.set.idx = (nrow(X) - length_test_dataset*3 + 1):nrow(X) 
```

and fit a Ridge Regression ($\alpha = 0$) with $\lambda = 0.75$ to the training dataset:

<div class="alert alert-block alert-danger">
<b>Beware of the wild parameter names</b> 

For both Ridge and Lasso Regression, we use the function `glm.net`. The $\alpha$ (alpha) parameter within this function describes the kind of penalty imposed on the weight:
    
   $\alpha = 0$ : Ridge Regression
    
   $\alpha = 1$ : Lasso Regression
    
The $\lambda$ (lambda) argument is the penalty on the weights just as described in the formulas in the introduction.

</div>

```r
lambda.ridge  = 0.75
our.ridge.fit = glmnet(x = X[-test.set.idx,],
                       y = Y[-test.set.idx],
                       family="gaussian", alpha=0, standardize=T, lambda = lambda.ridge)
```

We calculate the R2 and the MSE for our independet test dataset to check the performance of our model:

```r
our.ridge.prediction = predict(our.ridge.fit, X[test.set.idx,], lambda.ridge)
sprintf("R2 value of prediction on test dataset: %.2f",
        cor(c(our.ridge.prediction), Y[test.set.idx])^2)
sprintf("MSE value of prediction on test dataset: %.2f",
        MSE(c(our.ridge.prediction), Y[test.set.idx]))
```

We repeat the same for the Lasso ($\alpha = 1$) with $\lambda = 0.75$:

```r
lambda.lasso = 0.75
our.lasso.fit = glmnet(x = X[-test.set.idx,],
                       y = Y[-test.set.idx],
                       family="gaussian", alpha=1, standardize=T, lambda = lambda.lasso)
```

Predict for our independent test dataset:

```r
our.lasso.prediction = predict(our.lasso.fit, X[test.set.idx,], lambda.lasso)
sprintf("R2 value of prediction on test dataset: %.2f",
        cor(c(our.lasso.prediction), Y[test.set.idx])^2)
sprintf("MSE value of prediction on test dataset: %.2f",
        MSE(c(our.lasso.prediction), Y[test.set.idx]))
```

```r
plot(Y[test.set.idx], our.lasso.prediction, col="lightseagreen",
     ylab="Prediction precipitation anomaly [mm/d]",xlab="Test values precipitation anomaly [mm/d]",
     ylim=range(c(our.lasso.prediction,our.ridge.prediction)))
points(Y[test.set.idx], our.ridge.prediction, col="orchid4"); grid()
legend("topleft",legend=c("Ridge prediction","Lasso prediction","1:1 line"),
       col=c("orchid4","lightseagreen","red"),pch=c("o","o","-"), bg="white")
abline(0,1,col="red",lwd=2)
```

As you might have wondered, we have set the regularisation parameter to some initial value for illustration purposes ($\lambda = 0.75$). To understand the effect of the parameter, we explore how different values for $\lambda \in \{10^{-8},10^{-7.9},...,10^{1.9},10^{2}\}$ change the resulting $\beta$ parameters.

```r
## Only use 50 predictors (to make the whole thing faster)
set.seed(23)
feature_samples <- sample(ncol(X),50)

## Define range of lambdas for which beta coefficients should be estiamted
lambda_exponent <- seq(-8,2,length.out=101)
lambda_vec      <- 10^lambda_exponent
                      
## Calculate beta estimates for different lambda values:
betas_ridge <- cv.glmnet(x = X[-test.set.idx,feature_samples], y = Y[-test.set.idx],
                                                      family="gaussian", alpha=0, standardize=T,
                                                      lambda = lambda_vec)$glmnet.fit$beta[,length(lambda_vec):1]
betas_lasso <- cv.glmnet(x = X[-test.set.idx,feature_samples], y = Y[-test.set.idx],
                                                      family="gaussian", alpha=1, standardize=T,
                                                      lambda = lambda_vec)$glmnet.fit$beta[,length(lambda_vec):1]
```

```r
## Plot Ridge vs Lasso estimates for different lambdas:
cols <- colorRampPalette(brewer.pal(n = 11, name = "Spectral"))(11) 
ylim_plot <- range(rbind(betas_lasso,betas_ridge))

par(mfrow=c(2,1))
matplot(x=lambda_exponent, t(betas_ridge), type="l", ylim=ylim_plot,
        xlab="Lambda parameter 10^x", ylab="Beta estimates", main="Ridge", col=cols)
matplot(x=lambda_exponent, t(betas_lasso), type="l", ylim=ylim_plot,
        xlab="Lambda parameter 10^x", ylab="Beta estimates", main="Lasso", col=cols)
```

Based on 10-fold cross-validation we estimate the mean CV error for different $\lambda$ parameter values and determine the optimal parameter $\lambda_{\text{opt}}$ for both Ridge and Lasso Regression.

```r
## Fit model with all predictors
ridge.model = cv.glmnet(x = X, y = Y, family="gaussian", alpha = 0, 
                        nfolds = 10, standardize = T, lambda=lambda_vec)
lasso.model = cv.glmnet(x = X, y = Y, family="gaussian", alpha = 1,
                        nfolds = 10, standardize = T, lambda=lambda_vec)
```

```r
## Plot mean cross-validated error for ridge model
par(mfrow=c(2,1))
plot(ridge.model$lambda, ridge.model$cvm, type="l", log="x",col="red",
     ylab="Mean cv error", xlab="lambda", main="Ridge",
     ylim=c(min(ridge.model$cvlo),max(ridge.model$cvup)))
polygon(c(ridge.model$lambda,rev(ridge.model$lambda)),
        c(ridge.model$cvlo,rev(ridge.model$cvup)),
        col=rgb(1,0,0,0.5),border=NA)
abline(v=ridge.model$lambda[which.min(ridge.model$cvm)],lty="dashed")
legend("topleft",legend=sprintf("Lambda opt = %.4f", ridge.model$lambda[which.min(ridge.model$cvm)]),
      lty="dashed")

## Plot mean cross-validated error for lasso model
plot(lasso.model$lambda, lasso.model$cvm, type="l", log="x",col="red",
     ylab="Mean cv error", xlab="lambda", main="Lasso",
     ylim=c(min(lasso.model$cvlo),max(lasso.model$cvup)))
polygon(c(lasso.model$lambda,rev(lasso.model$lambda)),
        c(lasso.model$cvlo,rev(lasso.model$cvup)),
        col=rgb(1,0,0,0.5),border=NA)
abline(v=ridge.model$lambda[which.min(lasso.model$cvm)],lty="dashed")
legend("topleft",legend=sprintf("Lambda opt = %.4f", lasso.model$lambda[which.min(lasso.model$cvm)]),
      lty="dashed")
```

Make prediction for independent test dataset using the optimal lambda parameter $\lambda_{\text{opt}}$:

```r
ridge.prediction = predict(ridge.model, X[test.set.idx,], "lambda.min")
lasso.prediction = predict(lasso.model, X[test.set.idx,], "lambda.min")
 

sprintf("R2 value of ridge prediction on test dataset: %.2f",
        cor(c(ridge.prediction), Y[test.set.idx])^2)
sprintf("MSE value of ridge prediction on test dataset: %.2f",
        MSE(c(ridge.prediction), Y[test.set.idx]))

sprintf("R2 value of lasso prediction on test dataset: %.2f",
        cor(c(lasso.prediction), Y[test.set.idx])^2)
sprintf("MSE value of lasso prediction on test dataset: %.2f",
        MSE(c(lasso.prediction), Y[test.set.idx]))
```

# Part 2: Cross-Validation

For now, we have split our data in two parts: one for training the model and one for testing it. This is a necessary step to avoid overfitting our model to the noise in the data. Cross-Validation exploits this concept in a more effective way: we iteratively split our training data into a training fold (say, 80% of the training data) and a test fold.

<img src="figures/cv.png" width=400 height=400 />

As we have discussed in the lecture, this splitting is usually done by assigning the datapoints randomly to the cross-validation folds. But since in geoscientific data, our data usually has spatial and temporal dependencies and the variables are often correlated, the cross-validation split becomes more complicated.

In the following, we fit a $k$-nearest neighbor regression model to a similar scenario as in Part I of the exercise. Over Central Europe, we again want to predict precipitation from sea level pressure anomalies at different grid points (not only the Zurich grid point as in Part 1!), but the dataset is a bit smaller. As predictors we have

1. the local pressure anomaly,
2. the pressure anomaly of a point located 4 degrees northwards, and
3. the pressure anomaly of a point located 4 degrees west

(i.e. implicitly assuming that most winter low pressure systems over Central Europe move from west to east, north to south, or a combination...).

For determining the optimal $k$ parameter (number of neighbours used for averaging, c.f. last exercise), we use 10-fold cross-validation with 

1. random splits 
2. block-wise splits based on years

After this first inital data exploration, we can start our cross-validation exercise! Firstly, we cross-validate with random splits:

```r
nobs    <- nrow(train.data)        ## Total number of observations (n rows of X)
nclass  <- 5                       ## Number of cross-validation splits
knn.seq <- c(c(1, 3, 5, 10), seq(20, 100, 10), seq(150, 300, 50))

## Assign random CV assignment to observat
crossclass <- sample(nclass, nobs, T)

## Prepare empty matrix to fill with CV-predictions (random splits)
knn_CVrandomout <- matrix(data=NA, nrow = nobs, ncol = length(knn.seq))

## Loop over the nclass=5 folds (random split)
for (i in 1:nclass) {
  indtrain     <- which(crossclass!=i)        ## Select indices of observations not in cv-fold (training)
  indvalidate  <- setdiff(1:nobs,indtrain)    ## Select indices of observations in cv-fold (validations)
    
  ## Get predictions at validation indices for different nearest neigbour parameter k:
  knn_CVrandomout[indvalidate,1:length(knn.seq)] <- sapply(X = knn.seq,
                                                           FUN=function(k) kknn(formula = TOTPREC ~ PSL + PSL_Nshift + PSL_Wshift,
                                                                                train = train.data[-indvalidate,],
                                                                                test  = train.data[indvalidate,],
                                                                                k = k, kernel = "rectangular")$fitted.values)
}
```

... and then by year:

```r
## Assign temporally coherent folds assignment to observations:
tvec       <- unique(train.data$year)
crossclass <- c(rep.row(1:nclass, n = length(tvec) / nclass))

## Prepare empty matrix to fill with CV-predictions (random splits)
knn_CVyearout <- matrix(data=NA, nrow = nobs, ncol = length(knn.seq))

## Loop over the nclass=5 folds (temporal split)
for (i in 1:nclass) {
  indvalidate <- which(train.data$year %in% tvec[crossclass==i]) ## Select indices of observations not in cv-fold (training)
  
  ## Get predictions at validation indices for different nearest neigbour parameter k:
  knn_CVyearout[indvalidate,1:length(knn.seq)] <- sapply(X = knn.seq,
                                                         FUN=function(k) kknn(formula = TOTPREC ~ PSL + PSL_Nshift + PSL_Wshift,
                                                                              train = train.data[-indvalidate,],
                                                                              test = train.data[indvalidate,],
                                                                              k = k, kernel = "rectangular")$fitted.values)
}
```

Now we compare the performance of the two cross-validations with MSE and R2 metrics. We plot how the error changes with the number of neighbors $k$ for the random split and the split-by-year.

```r
## Calculate mean squared error MSE:
MSE_knn_CVrandomout <- sapply(X = 1:length(knn.seq), FUN=function(k.idx) MSE(obs = train.data$TOTPREC,
                                                                             pred = knn_CVrandomout[,k.idx]))
MSE_knn_CVyearout   <- sapply(X = 1:length(knn.seq), FUN=function(k.idx) MSE(obs = train.data$TOTPREC,
                                                                             pred = knn_CVyearout[,k.idx]))

## Calculate R^2 statistic:
R2_knn_CVrandomout <- sapply(X = 1:length(knn.seq), FUN=function(k.idx) cor(train.data$TOTPREC,
                                                                            knn_CVrandomout[,k.idx])^2)
R2_knn_CVyearout   <- sapply(X = 1:length(knn.seq), FUN=function(k.idx) cor(train.data$TOTPREC,
                                                                            knn_CVyearout[,k.idx])^2)

par(mfrow=c(1,2))
plot(c(1,1), ylim=c(0, 0.5), xlim = c(350, 1), type='n', xlab = "k - Nearest Neighbors", ylab = "R2", lwd = 2, xaxt="n", log="x")
grid()
axis(side = 1, at = knn.seq, labels = knn.seq)
lines(x = knn.seq, R2_knn_CVrandomout, lwd = 2, lty = 2)
lines(x = knn.seq, R2_knn_CVyearout, lwd = 2, lty = 1)
legend("bottom", c("Random split", "Leave-years-out"), lty = c(2, 1), lwd = 2, bg="white")
                           
plot(c(1,2), ylim=c(0.4, 1), xlim = c(350, 1), type='n', xlab = "k - Nearest Neighbors", ylab = "MSE [(mm/d)^2]", lwd = 2, xaxt="n", log="x")
grid()
axis(side = 1, at = knn.seq, labels = knn.seq)
lines(x = knn.seq, MSE_knn_CVrandomout, lwd = 2, lty = 2)
lines(x = knn.seq, MSE_knn_CVyearout, lwd = 2, lty = 1)
legend("topleft", c("Random split", "Leave-years-out"), lty = c(2, 1), lwd = 2, bg="white")
```
