# Part 1: Bias-Variance Trade-off

In the first part of today's exercise, we familiarise ourselves with a basic and very important concept of statistical learning, that has been introduced in the last lecture: the __Bias-Variance trade-off__.

No data is perfect. In fact, in statistical learning we usually assume the data is a mixture of the signal that interests us (e.g. rainfall at a specific location) and measurement noise (e.g. due to observational errors) that adds to it. Typically we would like to find a model that fits the signal and ignores the noise. This is where the Bias-Variance trade-off comes in:

If we start fitting a model to data, we have to somehow control the models flexibility. If the flexibility is too high, the model might be fitted to the noise potentially yielding low accuracy if applied to new data (i.e. high variance of the estimator). Too little flexibility imposes a systematic bias as the model cannot "follow" the data, i.e. it is not flexible enough to follow the signal correctly. The Bias-Variance trade-off arises along this gradient of model flexibility: In case of too much complexity, we fit the noise and get a high variance in our model (over-fitting). In case of too little complexity, we do not fit the signal enough and end up with a lot of bias (under-fitting).

The Bias-Variance trade-off is a basic concept that rearises in any statistical learning method that we will learn in this course. In the following, we will explore the Bias-Variance trade-off in some exemplary model fits.

## k-nearest neighbor regression vs. polynomial regression

We will now develop an understanding of the Bias-Variance tradeoff by fitting different models. For that purpose we use an artificial dataset ($Y$,$X$) with one predictor ($p = 1$) and $N = 200$ samples each in a
training and test dataset. We will fit

1. non-parametric $k$-nearest neighbor models of the form $ \hat{Y}(x) = \dfrac{1}{k} \sum_{x_i \in N_{k(x)}} y_i$ with varying $k$, ($k$ in $1..99$)$^1$, and
2. polynomial regression models $y_i = \beta_1 x_i + \beta_2 x_i^2 + ... \beta_q x_i^q + \epsilon_i$ with varying $q$, ($q$ in $1..25$).

$^1$ where we define $N_{k(x)}$ as the neighborhood of point $x$ with a specific $k$

We first load required packages and load an R Script of required functions for generating example data and fitting the models.

````r
## Load R-library 'kknn' and load R-functions in 'functions.R' file:
require("kknn")
source("functions.R")
````

We generate polynomial artifical training and test data with function `generate.data.poly` The artificial training data describes a polynomial function with added noise to it.

````r
## Generate 200 train and test data points
## (with residual standard deviation 0.15):
train.data = generate.data.poly(N=200, seed=200, sd.noise=0.15)
test.data  = generate.data.poly(N=200, seed=202, sd.noise=0.15)
````

This data describes a noisy polynomial function, which you can easily see when visualising the data.

````r
## Plot the function and the training and testing datapoints:
plot(train.data$X, train.data$Y_true, lwd=2, type="l", ylim=c(-0.5,1))
points(train.data$X, train.data$Y)
points(test.data$X, test.data$Y, col="red")
legend("topleft",legend=c("True","Train","Test"),col=c("black","black","red"),
       lty=c(1,NA,NA),pch=c(NA,1,1),lwd=2, bty = "n")
````

We then fit the $k$-nearest neighbour regression model for various $k$’s using the function `knnreg.fit` (similar to lecture). We calculate mean squared error for training and test dataset for all k’s using predefined function `MSE`.

<div class="alert alert-block alert-success">


````r
## Fit knn regression with k-coefficients 1 to 99:
knn.fits = lapply(X = 1:99, FUN=function(k) knnreg.fit(train.data = train.data, test.data = test.data, k = k))

## Get the mean squared error (MSE) w.r.t training and testing
## data points for all these 99 fits:
knn.mse.train = sapply(knn.fits, FUN=function(x) MSE(train.data$Y, x$Y.train))
knn.mse.test = sapply(knn.fits, FUN=function(x) MSE(test.data$Y, x$Y.train))
````

We plot the fitted k-nearest neighbour regression model for $k \in {1, 10, 50}$ and inspect how the fitted functions approximate our noisy training dataset.

````r
## Plot the predictions of three different knn fits:
par(mar=c(4,4,1,1))
plot(train.data$X, train.data$Y, pch = 16, xlab = "X", ylab = "Y", xlim = c(-1, 16), cex=1)
lines(train.data$X-c(diff(train.data$X)/2,0), knn.fits[[1]]$Y.train, col = "darkgray", lwd = 1.5, type="s")
lines(train.data$X-c(diff(train.data$X)/2,0), knn.fits[[10]]$Y.train, col = "orange", lwd = 1.5, type="s")
lines(train.data$X-c(diff(train.data$X)/2,0), knn.fits[[50]]$Y.train, col = "darkblue", lwd = 1.5, type="s")
legend("topleft", c("k-nearest neighbours - Training Set"), bty = "n")
legend("bottomright", c("k = 1", "k = 10", "k = 50"),
       col = c("darkgray", "darkorange", "darkblue"),
       lwd = 2, title = "k-nearest neighbours", bty = "n")
````

Now we switch to the polynomial regression model of various degrees $q$ and use it to fit our data (use function `poly.fit()` - or write your own function). We calculate the mean squared error for training and test dataset for all $q$’s up to $q=25$.

````r
## Fit polynomial regression with q-coefficients 1 to 25:
poly.fits = lapply(X = 1:25, FUN=function(degree) poly.fit(train.data = train.data, test.data = test.data,
                                                           degree = degree))
              
## Get the mean squared error (MSE) w.r.t training and testing
## data points for all these 25 fits:                                             
poly.fit.mse.train = sapply(poly.fits, FUN=function(x) MSE(train.data$Y, x$Y.train))
poly.fit.mse.test = sapply(poly.fits, FUN=function(x) MSE(test.data$Y, x$Y.train))                          
````

We plot the fitted polynomial regression model for $q \in \{1, 6, 25\}$ and inspect how the fitted functions approximate our noisy training dataset.

````r
## Plot the predictions of three different polynomial fits:
par(mar=c(4,4,1,1), mfrow=c(1, 1))
plot(train.data$X, train.data$Y, pch = 16, xlab = "X", ylab = "Y", xlim = c(-1, 16), cex=1)
lines(train.data$X, poly.fits[[1]]$Y.train, col = "darkgray", lwd = 2)
lines(train.data$X, poly.fits[[6]]$Y.train, col = "orange", lwd = 2)
lines(train.data$X, poly.fits[[25]]$Y.train, col = "darkblue", lwd = 2)
legend("topleft", c("Polynomial Regression - Training Set"), bty = "n")
legend("bottomright", c("q = 1", "q = 6", "q = 25"), col = c("darkgray", "darkorange", "darkblue"),
       lwd = 2, title = "Degree of polynomial", bty = "n")
````

We have seen examples of the Bias-Variance trade-off in a $k$-nearest neighbor regression and a polynomial regression exemplarly. We can further illustrate it by plotting how the performance of the models measured as mean square error, scales with the models' degrees of freedom ($k$ and $q$, respectively). In other words, we inspect the added value of model complexity for predicting the test and the training dataset.

````r
## Illustrate the Bias-Variance trade-off for k-nearest neighbor and polynomial regression.

## Plot bias-variance trade-off for knnreg:
par(mar=c(4,4,4,1), mfrow=c(1,2))
plot(knn.mse.train, xlim = c(100, 1), type='l', xlab = "Degrees of Freedom - N/k",
     ylab = "MSE", lwd = 2, xaxt="n", log="x") #, ylim=c(0, 10^6)
grid()
axis(side=1, at = c(1:100)[seq(1, 100, length.out=20)],
     labels = round(c(200 / 1:100)[seq(1, 100, length.out = 20)], 0))
axis(side=3, at = c(1:100)[seq(1, 100, length.out=20)],
     labels = round(c(1:100)[seq(1, 100, length.out = 20)], 0))
mtext(text = "k - Number of Nearest Neighbors", side = 3, line = 2)
lines(knn.mse.test, col = "darkorange", lwd = 2)
legend("topright", c("Training MSE", "Test MSE"), col = c("black", "darkorange"),
       lwd = 2, title = "k-nearest neighbours", bty = "n")

## Plot bias-variance trade-off for polynomial regression:
plot(c(1,0), xlim = c(0, 25), type='n', xlab = "Number of polynomials", ylab = "MSE", lwd = 2, ylim=c(0, 0.13))
grid()
lines(poly.fit.mse.train, col = "black", type="o", pch=16, cex=0.5)
lines(poly.fit.mse.test, col = "darkorange", type="o", pch=16, cex=0.5)
legend("topright", c("Training MSE", "Test MSE"), col = c("black", "darkorange"),
       lwd = 2, title = "Polynomial regression", bty = "n")
````

## Bias-Variance decomposition

In the lecture we have learned that the mean squared error can be decomposed:

> Expected test MSE = 
 
> = Bias + Variance + Irreducible Error

> = Var$(\hat{f}(x_0))$ + [Bias$(\hat{f}(x_0))]^2$ + Var$(\epsilon)$

Suppose you are given a set of 1000 independent realizations (e.g. ensemble members of a climate model) of 
an unknown function $f$, yielding $Y$ values ($N$=200) based on a set of fixed $X$ values.

a) Fit $k$-nearest neighbor regression models for $k \in \{1, 5, 10, 15, 20, 50, 100, 199\}$, and 

b) include estimates of [Bias$(\hat{f})]^2$, Var$(\hat{f})$, and Var$(e)$ for each $k$ in a plot of $k$ versus test MSE.

Hints:
- Generate data through: `data.large = lapply(X = 1:1000, FUN = function(idx) generate.data.poly(N = 200, seed = round(runif(1) * 10^5, 0)))`
- Calculate the true $f(X)$ directly from the large dataset (`ensemble mean'),
and $\mathop{{}\mathbb{E}}(\hat{f}(X))$ based on the fitted model.

````r
## Generate 1000 samples (first lapply-function) for each of the k-coefficients in k.seq:
data.large = lapply(X = 1:1000, FUN = function(idx) generate.data.poly(N = 200, seed = round(runif(1) * 10^5, 0), sd.noise = 0.15))

## Fit knn regression with different k-coefficients: 
k.seq = c(1, 2, 4, 5, 7, 10, 15, 20, 50, 100, 199)  
knnreg = lapply(X = 1:1000, FUN = function(idx) sapply(X = k.seq, FUN=function(k) knnreg.fit(train.data = data.large[[idx]], test.data = X, k = k)$Y.train))
````

````r
## Compare means over all 1000 samples with true underlying function:
f.true.samplmean = rowMeans(sapply(X = 1:1000, FUN=function(idx) data.large[[idx]]$Y))
f.true = train.data$Y_true
plot(f.true, f.true.samplmean,main="Scatterplot between 'true' function values\nand sample means")
abline(a=0,b=1,col="red")
````

We see that the sample mean is very close to the true function (not surprising, since we're calculating a the mean over 1000 i.i.d. samples).

````r
## Get training and test errors:
Emse.training = sapply(X = 1:length(k.seq), FUN=function(k.idx) mean(sapply(X = 1:1000, FUN=function(idx) MSE(knnreg[[idx]][,k.idx], data.large[[idx]]$Y))))
Emse.test = sapply(X = 1:length(k.seq), FUN=function(k.idx) mean(sapply(X = 1:999, FUN=function(idx) MSE(knnreg[[idx]][,k.idx], data.large[[idx+1]]$Y))))

## Get estimate of true function from mean over all 1000 knn fits (not samples!):                                                                        
E.fhat = sapply(X = 1:length(k.seq), FUN=function(k.idx) rowMeans(sapply(X = 1:1000, FUN=function(idx) knnreg[[idx]][,k.idx])))

## Get bias between mean E.fhat and f.true: 
E.bias.fhat = colMeans((E.fhat - f.true)^2)
                                                                         
## Get variance between E.fhat and f.true:
E.var.fhat = sapply(X = 1:length(k.seq), FUN=function(k.idx) mean(sapply(X = 1:1000, FUN = function(idx) (knnreg[[idx]][,k.idx])^2 - (E.fhat[,k.idx])^2)))
                           
## Get the residual variance:
var.epsilon = Emse.test - E.bias.fhat - E.var.fhat
````

````r
## Plot bias variance trade-off with independent components:
par(mar=c(4,4,4,1), mfrow=c(1,1))
plot(k.seq, Emse.training, xlim = c(200, 1), type='l', xlab = "Degrees of Freedom - N/k",
     ylab = "MSE", lwd = 2, xaxt="n", log="x") #, ylim=c(0, 10^6)
grid()
axis(side=1, at = k.seq, labels = round(200 / k.seq, 0))
axis(side=3, at = k.seq, labels = round(k.seq, 0))
mtext(text = "k - Number of Nearest Neighbors", side = 3, line = 2)
lines(k.seq, Emse.test, col = "darkorange", lwd = 2)
lines(k.seq, E.bias.fhat, col = "darkgray", lty =2, lwd = 2)
lines(k.seq, E.var.fhat, col = "darkgreen", lty =2, lwd = 2)
lines(k.seq, var.epsilon, col = "darkblue", lty =2, lwd = 2)

legend("topright", c("Training MSE", "Test MSE"), col = c("black", "darkorange"),
       lwd = 2, title = "k-nearest neighbours", bty = "n")
legend("right", c("??", "??", "??"), col = c("darkblue", "darkgray", "darkgreen"),
       lwd = 2, lty = 2, title = "Bias-Variance Tradeoff", bty = "n")
````

# Part 2: The special properties of Atmospheric and Climate Data

For the first part of the tutorial, we limited ourselves to one-dimensional, artifically created data. In the second part, we will now look into examples of real atmospheric and climate data, look at their most prominent features and their specific properties. We will discuss how this affects applying statistical learning methods to it.

__1) Complex data shape:__ Datasets describing variables important for processes in the atmosphere are generally not one-dimensional, but a multidimensional cube, often having a temporal dimension, two to three spacial dimensions (latitude, longitude, altitude), different variables (temperature, precipitation, ...) or spectral properties (in remote sensing) and so on. Most statistical learning methods, besides certain tailored deep learning algorithms like CNNs or LSTMs, need a 2D data format. It is therefore necessary to break down the N-dimensional data cube into a shape that can be digested by statistical learning methods, and how to do this is mostly not immediately obvious.

__2) Missing values:__ Earth observations are usually speckled with missing values. Observations may be subject to instrument failure, be discontinued or can only measure under certain conditions. For example satellite remote sensing datasets do only overpass certain regions at certain times, leaving large gaps of missing values. Statistical learning methods are typically not designed to deal with missing values. Usually, some sort of sensitive gapfilling needs to applied before applying statistical learning methods.

__3) Correlation:__ Statistical methods often assume independently sampled data. Atmospheric and Climate data expose a huge variety of correlations, e.g. between variables or between different places in space and time. Seasonality dominates many observations, thus e.g. the standard deviation of daily temperature is not representative for a stochastic element in daily temperature.

## Exploratory Data Analysis

Exploratory data analysis usually describes the first step of data analysis, where we try to understand what the main statistical and physical features of the data are. ## Autocorrelation

<div class="alert alert-block alert-success">
    <b>The acf( ) / pacf( ) functions</b>
    
Autocorrelation is a concept from time series analysis, and denotes the correlation between data points at different points in time (for example, how much is today's temperature in Zurich correlated with temperature in Zurich three days ago).
    
The ```acf()``` function computes (and by default plots) estimates of the autocorrelation function
    $\hat{\rho}(k)=\frac{\sum_{s=1}^{n-k}(x_{s+k}-\bar{x})(x_{s}-\bar{x})}{\sum_{t=1}{n}(x_{t}-\bar{x})^2}$, for $k=1, ..., n-1$, where $n$ is the is the time series length, and $k$ is the 'lag' between to data points (1, 2, ..) in a (regularly spaced) time series. With increasing lag $k$, we expect the correlation $\hat{\rho}(k)$ to go to zero (right?).
    
The partial autocorrelation function ```pacf()``` considers the fact that even if there only exists a linear dependence between data points with lag $k=1$, correlation will propagate to distant data points. Thus the ```pacf()``` function computes the autocorrelation between $X_t$ and $X_{t+k}$ with the linear dependence of $X_{t+1}$ through $X_{t+k-1}$ removed.
        
The blue dashed lines ( <font color='blue'>- - -</font> ) mark 95% confidence intervals, marking which correlations are still 'significantly' larger than zero.
</div>

````r
## Check for autocorrelation in the absolute temperature data:
par(mfrow=c(2,2))
acf(TREFHT_ZH_ts, main="ACF absolute temperature")
pacf(TREFHT_ZH_ts, main="PACF absolute temperature")
acf(TREFHT_ZH_decomp$time.series[,3], main="ACF residual temperature")
pacf(TREFHT_ZH_decomp$time.series[,3], main="PACF residual temperature")
````

<div class="alert alert-block alert-success">
    <b>glm( ) / glmnet( ) function</b>
    
The ```glm()``` function let's you fit an entire class of so-called '[generalized linear models](https://en.wikipedia.org/wiki/Generalized_linear_model)', whereof the standard linear regression with Gaussian residuals is one member (with option ```family = "gaussian"``` in the function call).
    
The ```glmnet()``` function is used to perform regularisation on the $\beta$-parameters, an option which can be 'deactivated by setting the option ```lambda=0``` in the function call. The concept of regularisation will be introduced next week, so stay tuned :)
</div>

````r
## Different fitting procedures, same result:
PSL_TOTPREC_df[,"dummy"] <- 0
PSL_TOTPREC_glm <- glm(TOTPREC ~ PSL, data=PSL_TOTPREC_df, family = "gaussian")
PSL_TOTPREC_reg <- glmnet(y=PSL_TOTPREC_df$TOTPREC, x=data.matrix(PSL_TOTPREC_df[,c("PSL","dummy")]),
                          family = "gaussian", alpha = 1, lambda=0)

print(c(  lm_beta1= PSL_TOTPREC_lm$coefficients[2],
         glm_beta1=PSL_TOTPREC_glm$coefficients[2],
  glmnet_beta1.PSL=PSL_TOTPREC_reg$beta[1]))
````
